```python
def evaluate_preds(train_true_values, train_pred_values, test_true_values, test_pred_values):
    print("Train R2:\t" + str(round(r2(train_true_values, train_pred_values), 3)))
    print("Test R2:\t" + str(round(r2(test_true_values, test_pred_values), 3)))
    
    plt.figure(figsize=(18,10))
    
    plt.subplot(121)
    sns.scatterplot(x=train_pred_values, y=train_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Train sample prediction')
    
    plt.subplot(122)
    sns.scatterplot(x=test_pred_values, y=test_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Test sample prediction')

    plt.show()
```

```python
# Шаг 3: Очистка данных и Feature engineering 

class DataPipeline:
    """Подготовка исходных данных"""
    
    def __init__(self):
        """Параметры класса"""
        self.medians = None
        self.kitchen_square_quantile = None
                
    def fit(self, X):
        """Сохранение статистик"""
       # self.medians=X['HouseYear','Rooms','Square'].median()
        self.medians = X.median()
        self.kitchen_square_quantile = X['KitchenSquare'].quantile(.98)
        
    def transform(self, X): 
    
        # Rooms
        X['Rooms_newcol'] = 0
        X.loc[(X['Rooms'] == 0) | (X['Rooms'] >= 5), 'Rooms_newcol'] = 1
        X.loc[X['Rooms'] == 0, 'Rooms'] = 1
        X.loc[X['Rooms'] >= 5, 'Rooms'] = self.medians['Rooms']
        
        # KitchenSquare
        X.loc[(X['Square']<X['LifeSquare'])|(X['KitchenSquare']<3), 'LifeSquare'] = X['Square']
        X.loc[X['KitchenSquare'] < 3, 'KitchenSquare'] = 0
        X.loc[X['Square']<X['KitchenSquare'], 'KitchenSquare']=X['Square']-X['LifeSquare']
        
        X['KitchenSquare'].quantile(.99), X['KitchenSquare'].quantile(.025)
        condition = (X['KitchenSquare'].isna()) \
                     | (X['KitchenSquare'] > X['KitchenSquare'].quantile(.99))
        X.loc[condition, 'KitchenSquare'] =self.medians['KitchenSquare']
        #X.loc[train_df['KitchenSquare'] < 3, 'KitchenSquare'] = 3

        # HouseFloor, Floor
        X.loc[X['HouseFloor'] == 0, 'HouseFloor'] = X['HouseFloor'].median()
        X.loc[X['HouseFloor']<X['Floor'], 'HouseFloor']=X['Floor'] 
        
        # HouseYear
        current_year = datetime.now().year
        #X=X.drop(X[X['HouseYear'] >= (current_year+3)].index)
        X.loc[X['HouseYear'] > current_year, 'HouseYear'] = current_year

        # Healthcare_1
        if 'Healthcare_1' in X.columns:
            X.drop('Healthcare_1', axis=1, inplace=True)
            
        # LifeSquare
        X['LifeSquare']=X['LifeSquare'].fillna(X['Square']-X['KitchenSquare'])
        
        return X
```

```python
#!
class FeatureGenetator():
    """Генерация новых фич"""
    
    def __init__(self):
        self.DistrictId_counts = None
        self.binary_to_numbers = None
        self.med_price_by_district = None
        self.med_price_by_year = None
        self.med_price_by_eco = None
        self.med_price_by_meter = None
        self.house_year_max = None
        self.district_size = None
        
        
    def fit(self, X, y=None):
        
        #X = X.copy()
        
        # Binary features
        self.binary_to_numbers = {'A': 0, 'B': 1}
        
        # DistrictID
        self.district_size = X['DistrictId'].value_counts().reset_index() \
                               .rename(columns={'index':'DistrictId', 'DistrictId':'DistrictSize_newcol'})
                
        ## District, Rooms
        df = X.copy()
        
        if y is not None:
            df['Price'] = y.values
            
            self.med_price_by_district = df.groupby(['DistrictId', 'Rooms'], as_index=False).agg({'Price':'median'})\
                                            .rename(columns={'Price':'MedPriceByDistrict_newcol'})
            
            self.med_price_by_district_median = self.med_price_by_district['MedPriceByDistrict_newcol'].median()
            
        ## year
        if y is not None:  
            self.house_year_max = df['HouseYear'].max()
            df['Price'] = y.values
            df = self.year_to_cat(df)
            self.med_price_by_year = df.groupby(['year_cat_newcol'], as_index=False).agg({'Price':'median'}).\
                                            rename(columns={'Price':'MedPriceByYear_newcol'})
            self.med_price_by_year_median = self.med_price_by_year['MedPriceByYear_newcol'].median()
            
            ## district, ecology
        if y is not None:
            df['Price'] = y.values 
            
            self.med_price_by_eco = df.groupby(['DistrictId', 'Ecology_1'], as_index=False).agg({'Price':'median'})\
                                            .rename(columns={'Price':'MedPriceByEco_newcol'})
            
            self.med_price_by_eco_median = self.med_price_by_eco['MedPriceByEco_newcol'].median()
            
            ## district, square_meter
        if y is not None:
            df['Price'] = y.values 
            
            df['Newcol'] = df['Price'] /df['Square']
            
            self.med_price_by_meter = df.groupby(['DistrictId'], as_index=False).agg({'Newcol':'median'})\
                                          .rename(columns={'Newcol':'MedPriceByMeter_newcol'})
            
            self.med_price_by_meter_median = self.med_price_by_meter['MedPriceByMeter_newcol'].median()
                
    def transform(self, X):
        
        # Binary features
        X['Ecology_2'] = X['Ecology_2'].map(self.binary_to_numbers)  # self.binary_to_numbers = {'A': 0, 'B': 1}
        X['Ecology_3'] = X['Ecology_3'].map(self.binary_to_numbers)
        X['Shops_2'] = X['Shops_2'].map(self.binary_to_numbers)
        
        # DistrictId, IsDistrictLarge
        X = X.merge(self.district_size, on='DistrictId', how='left')
        
        X['new_district'] = 0
        X.loc[X['DistrictSize_newcol'].isna(), 'new_district'] = 1        
        X['DistrictSize_newcol'].fillna(5, inplace=True)        
        X['IsDistrictLarge_newcol'] = (X['DistrictSize_newcol'] > 100).astype(int)
        
        # More categorical features
        X = self.year_to_cat(X)   # + столбец year_cat
        
        # Target encoding
        if self.med_price_by_district is not None:
            X = X.merge(self.med_price_by_district, on=['DistrictId', 'Rooms'], how='left')
            X['MedPriceByDistrict_newcol'].fillna(self.med_price_by_district_median, inplace=True)
            
        if self.med_price_by_year is not None:
            X = X.merge(self.med_price_by_year, on=['year_cat_newcol'], how='left')
            X['MedPriceByYear_newcol'].fillna(self.med_price_by_year_median, inplace=True)
            
        if self.med_price_by_eco is not None:
            X = X.merge(self.med_price_by_eco, on=['DistrictId', 'Ecology_1'], how='left')
            X['MedPriceByEco_newcol'].fillna(self.med_price_by_eco_median, inplace=True)
            
        if self.med_price_by_meter is not None:
            X = X.merge(self.med_price_by_meter, on=['DistrictId'], how='left')
            X['MedPriceByMeter_newcol'].fillna(self.med_price_by_meter_median, inplace=True)
        
        return X
    
    def year_to_cat(self, X):
        bins = [0, X['HouseYear'].quantile(.25), X['HouseYear'].quantile(.5), X['HouseYear'].quantile(.75), X['HouseYear'].quantile(.9),  X['HouseYear'].max()]
        
        X['year_cat_newcol'] = pd.cut(X['HouseYear'], bins=bins, labels=False)

        X['year_cat_newcol'].fillna(-1, inplace=True)
        return X
    
```

```python
#!
feature_names = ['Rooms', 'Square', 'LifeSquare', 'KitchenSquare', 'Floor', 'HouseFloor', 'HouseYear',
                 'Ecology_1', 'Ecology_2', 'Ecology_3', 'Social_1', 'Social_2', 'Social_3',
                 'Helthcare_2', 'Shops_1', 'Shops_2']

new_feature_names = ['DistrictSize_newcol', 'new_district', 'IsDistrictLarge_newcol', 'MedPriceByMeter_newcol', 'MedPriceByEco_newcol', 'MedPriceByDistrict_newcol', 'MedPriceByYear_newcol']

target_name = 'Price'
```

```python
#Предобработка данных
preproc = DataPipeline()
preproc.fit(train_df)
train_df = preproc.transform(train_df)

# Разбиение на train и test
X = train_df.drop(columns=target_name)
y = train_df[target_name]

X_train, X_valid, Y_train, Y_valid = train_test_split(X, y, test_size=0.33, shuffle=True, random_state = 21)
#Предобработка данных
#preproc = DataPipeline()
#preproc.fit(X_train)
#X_train = preproc.transform(X_train)
#X_valid = preproc.transform(X_valid)


X_train.shape, X_valid.shape, test_df.shape, Y_train.shape

#Предобработка данных
preproc = DataPipeline()
preproc.fit(X_train)
X_train = preproc.transform(X_train)
X_valid = preproc.transform(X_valid)
test_df = preproc.transform(test_df)

#X_train.shape, X_valid.shape, test_df.shape, Y_train.shape
```

```python
# Создаем новые фичи

features_gen = FeatureGenetator()
features_gen.fit(X_train, Y_train)

X_train = features_gen.transform(X_train)
X_valid = features_gen.transform(X_valid)
test_df = features_gen.transform(test_df)

X_train.shape, X_valid.shape, test_df.shape
```

```python
X_train = X_train[feature_names + new_feature_names]
X_valid = X_valid[feature_names + new_feature_names]
test_df = test_df[feature_names + new_feature_names]
```
